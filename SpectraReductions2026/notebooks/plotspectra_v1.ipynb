```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import os
from dotenv import load_dotenv
PROJECT_ROOT = Path(os.getenv("PROJECT_ROOT", Path.cwd())).resolve()
if not (PROJECT_ROOT / "README.md").exists() and (Path.cwd().parent / "README.md").exists():
    PROJECT_ROOT = Path.cwd().parent.resolve()
load_dotenv(PROJECT_ROOT / ".env", override=True)

# ------------------------------------------------------------
# Files
# ------------------------------------------------------------
base = Path(os.getenv("NORMALIZED_AAT_DIR", "/Users/kerrycheon/repos/SpectraReductions2026/data/interim/aat_normalized_lab_frame")).resolve()

best_file = base / "NGC6569_best_averages_alpha_rule_full.csv"
bes_file  = base / "besan_6569.csv"

df  = pd.read_csv(best_file)
bes = pd.read_csv(bes_file)

# ------------------------------------------------------------
# Choose sample
#   sample_mode = "all"  -> all fibers with finite best-rule values
#   sample_mode = "rv"   -> only -65 <= RV_barycentric <= -25 km/s
# ------------------------------------------------------------
sample_mode = "rv"

# ------------------------------------------------------------
# Build selections
# ------------------------------------------------------------
all_fibers = df[
    np.isfinite(df["FeH_for_TW_rule"]) &
    np.isfinite(df["aFe_for_TW_rule"]) &
    np.isfinite(df["RV_barycentric"])
].copy()

rv_selected = all_fibers[
    (all_fibers["RV_barycentric"] >= -65.0) &
    (all_fibers["RV_barycentric"] <= -25.0)
].copy()

if sample_mode == "rv":
    plot_df = rv_selected.copy()
    title = "Tinsley–Wallerstein Diagram: RV-Selected Objects"
    out_stub = "Tinsley_Wallerstein_best_rule_rv_selected_color_RV"
else:
    plot_df = all_fibers.copy()
    title = "Tinsley–Wallerstein Diagram: All Fibers"
    out_stub = "Tinsley_Wallerstein_best_rule_all_fibers_color_RV"

print("All fibers with finite best-rule pair and RV:", len(all_fibers))
print("RV-selected fibers:", len(rv_selected))
print("Plotting sample:", len(plot_df))

# ------------------------------------------------------------
# Helper to add average error bar
# ------------------------------------------------------------
def add_average_errorbar(ax, data, xerr_col="e_FeH_adopted", yerr_col="e_aFe_best_rule",
                         x0=-1.38, y0=-0.20):
    xerr_vals = pd.to_numeric(data[xerr_col], errors="coerce")
    yerr_vals = pd.to_numeric(data[yerr_col], errors="coerce")

    mean_xerr = np.nanmean(xerr_vals)
    mean_yerr = np.nanmean(yerr_vals)

    print(f"Average {xerr_col}: {mean_xerr:.3f}")
    print(f"Average {yerr_col}: {mean_yerr:.3f}")

    ax.errorbar(
        x0, y0,
        xerr=mean_xerr,
        yerr=mean_yerr,
        fmt="none",
        ecolor="black",
        elinewidth=1.2,
        capsize=4,
        capthick=1.2,
        zorder=20
    )

# ------------------------------------------------------------
# Plot
# ------------------------------------------------------------
fig, ax = plt.subplots(figsize=(8.6, 5.6))

# Besancon background
ax.scatter(
    bes["[M/H]"],
    bes["[a/Fe]"],
    s=8,
    alpha=0.12,
    color="gray",
    linewidths=0,
    zorder=1
)

# Science points, color-coded by barycentric RV
sc = ax.scatter(
    plot_df["FeH_for_TW_rule"],
    plot_df["aFe_for_TW_rule"],
    c=plot_df["RV_barycentric"],
    cmap="viridis_r",      # reverse so more negative RVs are darker if you like
    s=70,
    edgecolors="black",
    linewidths=0.8,
    zorder=10
)

# Optional: plot cluster reference mean as a black star
ax.scatter(
    [-0.83], [0.38],
    marker="*",
    s=220,
    color="black",
    zorder=15
)

# Average error bar
add_average_errorbar(
    ax,
    plot_df,
    xerr_col="e_FeH_adopted",
    yerr_col="e_aFe_best_rule",
    x0=-1.40,
    y0=-0.20
)

# Axes and labels
ax.set_xlabel(r"[Fe/H]")
ax.set_ylabel(r"[$\alpha$/Fe]")
ax.set_title(title)
ax.set_xlim(-1.65, 0.50)
ax.set_ylim(-0.40, 0.90)
ax.grid(alpha=0.25)

# Colorbar
cbar = plt.colorbar(sc, ax=ax)
cbar.set_label(r"$RV_{\rm barycentric}$ (km s$^{-1}$)")

plt.tight_layout()

# Save
png_out = base / f"{out_stub}.png"
pdf_out = base / f"{out_stub}.pdf"

plt.savefig(png_out, dpi=250, bbox_inches="tight")
plt.savefig(pdf_out, bbox_inches="tight")
plt.show()

print("Saved:")
print(png_out)
print(pdf_out)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import os
from dotenv import load_dotenv
PROJECT_ROOT = Path(os.getenv("PROJECT_ROOT", Path.cwd())).resolve()
if not (PROJECT_ROOT / "README.md").exists() and (Path.cwd().parent / "README.md").exists():
    PROJECT_ROOT = Path.cwd().parent.resolve()
load_dotenv(PROJECT_ROOT / ".env", override=True)

# ------------------------------------------------------------
# Input files
# ------------------------------------------------------------
base = Path(os.getenv("NORMALIZED_AAT_DIR", "/Users/kerrycheon/repos/SpectraReductions2026/data/interim/aat_normalized_lab_frame")).resolve()

aat_file = base / "NGC6569_best_averages_alpha_rule_full.csv"
bes_file = base / "besan_6569.csv"

aat = pd.read_csv(aat_file)
bes = pd.read_csv(bes_file)

# ------------------------------------------------------------
# Complete sample
# ------------------------------------------------------------
aat = aat[
    np.isfinite(aat["radeg"]) &
    np.isfinite(aat["decdeg"]) &
    np.isfinite(aat["RV_barycentric"])
].copy()

if "current_paper_style_candidate" in aat.columns:
    cand = aat[aat["current_paper_style_candidate"] == True].copy()
else:
    cand = aat.iloc[0:0].copy()

ra0 = 273.412
dec0 = -31.827
rt_arcmin = 6.9

print(f"AAT complete sample: {len(aat)}")
print(f"Current paper-style candidates: {len(cand)}")
print(f"Besançon comparison stars: {len(bes)}")

# ------------------------------------------------------------
# Figure layout
# map | colorbar | spacer | histogram
# ------------------------------------------------------------
fig = plt.figure(figsize=(13.6, 5.9))

gs = fig.add_gridspec(
    1, 4,
    width_ratios=[1.00, 0.05, 0.18, 1.10],
    wspace=0.14
)

ax1 = fig.add_subplot(gs[0, 0])
cax = fig.add_subplot(gs[0, 1])
spacer = fig.add_subplot(gs[0, 2])
ax2 = fig.add_subplot(gs[0, 3])

spacer.axis("off")

# ------------------------------------------------------------
# Panel (a): sky distribution
# ------------------------------------------------------------
sc = ax1.scatter(
    aat["radeg"],
    aat["decdeg"],
    c=aat["RV_barycentric"],
    cmap="turbo",
    s=35,
    linewidths=0.35,
    edgecolors="0.35",
    alpha=0.95,
    zorder=3
)

if len(cand) > 0:
    ax1.scatter(
        cand["radeg"],
        cand["decdeg"],
        marker="*",
        s=160,
        facecolors="#9cff57",
        edgecolors="black",
        linewidths=0.8,
        zorder=6
    )

ax1.scatter(
    [ra0], [dec0],
    marker="X",
    s=260,
    color="#1f77b4",
    edgecolors="black",
    linewidths=1.0,
    zorder=7
)

theta = np.linspace(0, 2*np.pi, 500)
cosdec = np.cos(np.deg2rad(dec0))

for i in range(1, 6):
    r = i * rt_arcmin
    dx_arcmin = r * np.cos(theta)
    dy_arcmin = r * np.sin(theta)

    ra_ring = ra0 + dx_arcmin / (60.0 * cosdec)
    dec_ring = dec0 + dy_arcmin / 60.0

    ax1.plot(
        ra_ring,
        dec_ring,
        color="gray",
        lw=0.9,
        alpha=0.55,
        zorder=1
    )

r = rt_arcmin
dx_arcmin = r * np.cos(theta)
dy_arcmin = r * np.sin(theta)

ra_ring = ra0 + dx_arcmin / (60.0 * cosdec)
dec_ring = dec0 + dy_arcmin / 60.0

ax1.fill(
    ra_ring,
    dec_ring,
    color="gray",
    alpha=0.10,
    zorder=0
)

ax1.set_xlabel("RA")
ax1.set_ylabel("Dec", labelpad=6)
#ax1.set_title("(a)", loc="bottom left", fontsize=14)
ax1.grid(alpha=0.25)

pad_ra = 0.05
pad_dec = 0.05
ax1.set_xlim(aat["radeg"].max() + pad_ra, aat["radeg"].min() - pad_ra)
ax1.set_ylim(aat["decdeg"].min() - pad_dec, aat["decdeg"].max() + pad_dec)

cbar = fig.colorbar(sc, cax=cax)
cbar.set_label(r"RV (km s$^{-1}$)", rotation=270, labelpad=20)

# ------------------------------------------------------------
# Panel (b): RV histogram
# ------------------------------------------------------------
aat_w = np.ones(len(aat)) / len(aat)
cand_w = np.ones(len(cand)) / len(cand) if len(cand) > 0 else None
bes_w = np.ones(len(bes)) / len(bes)

rv_min = min(float(aat["RV_barycentric"].min()), float(bes["HRV"].min()))
rv_max = max(float(aat["RV_barycentric"].max()), float(bes["HRV"].max()))

bin_width = 15.0
bins = np.arange(
    np.floor(rv_min / bin_width) * bin_width,
    np.ceil(rv_max / bin_width) * bin_width + bin_width,
    bin_width
)

ax2.hist(
    bes["HRV"],
    bins=bins,
    weights=bes_w,
    alpha=0.35,
    color="#8da0ff",
    label="Besançon model"
)

ax2.hist(
    aat["RV_barycentric"],
    bins=bins,
    weights=aat_w,
    histtype="step",
    linewidth=2.0,
    color="#ff7f0e",
    label="AAT complete sample"
)

if len(cand) > 0:
    ax2.hist(
        cand["RV_barycentric"],
        bins=bins,
        weights=cand_w,
        alpha=0.85,
        color="#2ca02c",
        label="AAT candidate set"
    )

ax2.set_xlabel(r"RV (km s$^{-1}$)")
ax2.set_ylabel("Relative Frequency", labelpad=8)
#ax2.set_title("(b)", loc="left", fontsize=14)
ax2.grid(alpha=0.25)
ax2.legend(frameon=False, fontsize=9, loc="upper right")
ax2.set_xlim(bins.min(), bins.max())
# Panel labels in bottom-left corners
ax1.text(
    0.03, 0.04, "(a)",
    transform=ax1.transAxes,
    fontsize=14,
    fontweight="bold",
    ha="left",
    va="bottom"
)

ax2.text(
    0.03, 0.04, "(b)",
    transform=ax2.transAxes,
    fontsize=14,
    fontweight="bold",
    ha="left",
    va="bottom"
)
# ------------------------------------------------------------
# Title and spacing
# ------------------------------------------------------------
fig.suptitle(
    "NGC 6569 Complete Spectroscopic Dataset: Sky Distribution and RV Histogram",
    y=0.97,
    fontsize=18
)

fig.subplots_adjust(
    left=0.06,
    right=0.98,
    bottom=0.13,
    top=0.90,
    wspace=0.14
)

# ------------------------------------------------------------
# Save
# ------------------------------------------------------------
png_out = base / "NGC6569_complete_dataset_map_histogram_more_spaced.png"
pdf_out = base / "NGC6569_complete_dataset_map_histogram_more_spaced.pdf"

fig.savefig(png_out, dpi=250, bbox_inches="tight")
fig.savefig(pdf_out, bbox_inches="tight")
plt.show()

print("Saved:")
print(png_out)
print(pdf_out)
```
